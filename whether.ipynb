# Make a dummy variable and stored all the data

```python
weather_dummies=pd.get_dummies(weather)
weather_dummies.head()
```

# First spilit the data 

```python
from sklearn.model_selection import train_test_split as tts
train, test=tts(weather_dummies,test_size=0.2,random_state=42)
```

```python
train_x=train.loc[:, train.columns !="RainTomorrow"]
test_x=test.loc[:, train.columns !="RainTomorrow"]

train_y=train["RainTomorrow"]
test_y=test['RainTomorrow']
```

# Using Decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(train_x, train_y)
```

```python
test_pred=model.predict(test_x)
```

```python
result=pd.DataFrame({'Actual Score':test_y,'predicted_score':test_pred})
print(result)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score as ac
print(classification_report(test_y,test_pred))
```

```python
print(confusion_matrix(test_y,test_pred))
```

```python
print(ac(test_y,test_pred))*100
```

# Using Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_jobs = 2)
clf.fit(train_x, train_y)
```

```python
test_pred1=clf.predict(test_x)
```

```python
result=pd.DataFrame({'Actual Score':test_y,'predicted_score':test_pred1})
print(result)
```

```python
print(classification_report(test_y,test_pred1))
```

```python
print(confusion_matrix(test_y,test_pred1))
```

```python
print(ac(test_y,test_pred1))*100
```
