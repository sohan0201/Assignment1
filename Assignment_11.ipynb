# importing libraries

```python
import numpy as np
import pandas as pd
```

# Loading the datasets and concatenate

```python
train=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')

test=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')

data=[train,test]
titanic=pd.concat(data)
titanic.head()
```

```python
titanic.shape
```

```python
titanic.index
```

```python
titanic.columns
```

# Find missing values

```python
titanic.isnull().sum()
```

# Remove Nan values into survied

```python
titanic['Survived']=titanic['Survived'].dropna()
```

```python
titanic['Survived'].value_counts()
```

```python
titanic['Has_Cabin']=titanic.Cabin.isnull()
```

```python
titanic
```

# Extract the Saluation from Name and stored in Title column using RE 

```python
import re
titanic['Title'] = titanic.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
```

```python
titanic
```

# Impute the missing values using .fillna() method.

```python
titanic.isnull().sum()
```

```python
titanic.Age=titanic.Age.fillna(titanic.Age.median())
titanic.Fare=titanic.Fare.fillna(titanic.Fare.median())
titanic.Embarked=titanic.Embarked.fillna('S')
titanic.tail()
```

```python
titanic.isnull().sum()
```

# Create two new columns fill with qcut in same bin

```python
titanic['Categorical_age']=pd.qcut(titanic.Age,q=4,labels=False)
titanic['Categorical_fare']=pd.qcut(titanic.Fare,q=4,labels=False)
titanic.head()
```

# Those columns are unuseful information drop it like Cabin, Name, PassengerId,Ticket

```python
titanic=titanic.drop(['Cabin','Name','PassengerId','Ticket'],axis=1)
titanic.head()
```

# Now we have all information of Age and Fare in bin so you can easily drop it

```python
titanic=titanic.drop(['Age','Fare'],axis=1)
titanic.head()
```

# Numbers of Members in Family Onboard

```python
titanic['Fam_size']=titanic.Parch+titanic.SibSp
titanic=titanic.drop(['Parch','SibSp'],axis=1)
titanic.head()
```

# Transform all variables into numeric ones using get_dummies().

```python
titanic_dummies=pd.get_dummies(titanic)
titanic_dummies.head()
```

# X(Features) will be the dataset values without the column â€œSurvivedâ€ and y(Targets) will be your â€œSurvivedâ€ column values.

```python
X = titanic_dummies.drop(['Survived'], axis = 1).values
y = titanic_dummies['Survived'].values
```

# Preparing the datasets

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.3, random_state=42)
```

#  Fitting the line using logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

# Predictions


```python
test_pred =lr.predict(X_test)
```

```python
result=pd.DataFrame({'Actual':y_test,'Predicted':test_pred})
print(result)
```

# Accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy_score = lr.score(X_test, y_test)
print("The accuracy score is : {}".format(accuracy_score))
```
